```python
def lnlike(theta,dat,cov,cinv):
    gauss_like = lambda thy: np.exp(-0.5*np.dot(np.dot(thy-dat,cinv),thy-dat)) / np.sqrt(np.linalg.det(2*np.pi*cov))
    return np.log(gauss_like(theta))
```

```python
def lnprior(theta):
    x,y = theta
    return 0.0
```

```python
def lnprob(theta):
    lp = lnprior(theta)
    if not lp == -np.inf:
        return -np.inf
    return lp + lnlike(theta)
```
